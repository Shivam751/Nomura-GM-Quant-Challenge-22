```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('IPL_train.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.groupby('match_id').sum()['total_runs']
```

```python
data['runs_now'] = data.groupby('match_id').cumsum()['total_runs']
```

```python
data['player_dismissed'].fillna(0, inplace = True)
```

```python
data['player_dismissed'] = data['player_dismissed'].apply(lambda x: 1 if type(x) == str else 0)
```

```python
data['wickets_now'] = data.groupby('match_id').cumsum()['player_dismissed']
```

```python
data.head()
```

```python
data['ball_done'] = (data['over'] - 1)*6 + data['ball'] 
```

```python
data.head()
```

```python
data['run_rate'] = data['runs_now'] / (data['ball_done']/6)
```

```python
data.head()
```

```python
df = data.filter(['match_id', 'batting_team', 'bowling_team', 'batsman', 'non_striker', 'bowler', 'runs_now', 'wickets_now', 'ball_done', 'run_rate'], axis = 1)
```

```python
df.head()
```

```python
df['match_id'].value_counts()
```

```python
total_runs = data.groupby('match_id').sum()['total_runs'].values
```

```python
idx_match_change = []
idx_match_change.append(-1)
for i in range(len(df)-1):
    if df['match_id'][i] != df['match_id'][i+1]:
        idx_match_change.append(i)
```

```python
idx_match_change.append(len(df)-1)
```

```python
df['score'] = np.zeros(len(df))
for i in range(1, len(idx_match_change)):
    for j in range(idx_match_change[i-1]+1, idx_match_change[i]+1):
        df['score'][j] = total_runs[i-1]
```

```python
df.to_csv('data_trans.csv', index=False)
```

```python
df.head()
```

```python
y = df['score']
X = df.filter(['runs_now', 'wickets_now', 'ball_done', 'run_rate'], axis = 1)
```

```python
X
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
regressor.fit(X, y)
```
